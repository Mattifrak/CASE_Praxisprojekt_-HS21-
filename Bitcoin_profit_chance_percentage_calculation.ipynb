```python
#Putting the 5 periods in a different variable
daily_period = 1
weekly_period = 7
monthly_period = 30
yearly_period = 365
two_yearly_period = 730
```

```python
#Make extra tables with price values shifted up n amount
bc['Daily_difference'] = bc[['Price']].shift(-daily_period)
bc['Weekly_difference'] = bc[['Price']].shift(-weekly_period)
bc['Monthly_difference'] = bc[['Price']].shift(-monthly_period)
bc['Yearly_difference'] = bc[['Price']].shift(-yearly_period)
bc['Two_yearly_difference'] = bc[['Price']].shift(-two_yearly_period)
```

```python
#Loop that puts a 1 in the empty table if price has increased or stayed the same after an n period. In case of decrease, it puts a 0
n = 0
for n in range(len(bc)):
  if bc['Daily_difference'].iloc[n]>=bc['Price'].iloc[n]:
    bc.iat[n, bc.columns.get_loc('Daily')] = 1
    
  else:
    bc.iat[n, bc.columns.get_loc('Daily')] = 0
  
```
